```python
#Создадим переменную для названия всех столбцов
df.columns
column_names = df.columns
```

```python
#Диаграмма ящик с усами
for column in df.columns:
    fig = px.box(df, y = column)
    fig.show()
```

```python
#Посчитаем какое кол-во у угла нашивки, равное 0
df['Угол нашивки, град'][df['Угол нашивки, град'] == 0.0].count()


```

```python
#Посчитаем какое кол-во у угла нашивки, равное 90
df['Угол нашивки, град'][df['Угол нашивки, град'] == 90.0].count()
```

```python
#Так как на графиках выше видно, что значения у угла нашивки равны 0 или 90, то возможно преобразовать эти значения в 0 и 1
df = df.replace({'Угол нашивки, град': {0.0 : 0, 90.0 : 1}})
df['Угол нашивки, град'] = df['Угол нашивки, град'].astype(int)
```

```python
#Проверим какое кол-во у угла нашивки, равное 0 после преобразования
df['Угол нашивки, град'][df['Угол нашивки, град'] == 0].count()
```

```python
#Проверим какое кол-во у угла нашивки, равное 1 после преобразования
df['Угол нашивки, град'][df['Угол нашивки, град'] == 1].count()
```

```python
#Еще раз выведем диаграмму ящик с усами
for column in df.columns:
    fig = px.box(df, y = column)
    fig.show()
```

```python
#Посмотрим на более компактном варианте ящик с усами
sns.boxplot(data = df, orient="h")
plt.show()
```

```python
#Попарные графики рассеяния точек
g = sns.PairGrid(df[df.columns])
g.map(sns.scatterplot, color = 'darkblue')
g.map_upper(sns.scatterplot, color = 'darkblue')
g.map_lower(sns.kdeplot, color = 'darkblue')
plt.show
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(16, 11))
sns.heatmap(corr, vmin=-0.5, vmax=0.5, annot=True, fmt='.2f',cmap='PRGn', ax=ax, linewidths = 0.1)
plt.show()
```
